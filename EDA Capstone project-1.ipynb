### Let's begin Visualizations!

```python
# we will be plotting the count of attacks according to the their respective years
df.groupby(['iyear']).size().plot(kind='bar',color='r')
plt.rcParams["figure.figsize"] = (15,10)
plt.xlabel('Year----->')
plt.ylabel('number of attacks------>')
```

### As it Could be seen, when we plot the count of terror attacks in their perspective attacks, 2014 stands on the top with a count of more than 16000 attacks in that year.

### The number of attacks gradually increase from 1970 till 2014 and we notice a gradual decreasing trend since.

```python
# plotting countries vs number of attacks 
df.groupby(['country_txt']).size().plot(kind='bar')
plt.rcParams["figure.figsize"] = (60,10)
plt.xlabel('Country Name----->')
plt.ylabel('number of attacks------>')
# plt.xticks(rotation=45)
```

### So it is of no surprise to see that the top 2 Countries where the attacks happened are Iraq and Pakistan!!. And the consequent places are taken by Afghanistan and India.

#### To See the results clearly as it is not that easy to make from the visualization, which country ranks first in terrorism, We will be using Groupby function from  the pandas library 

```python
df_new=df.groupby(['country_txt']).count()
```

```python
# Grouping by country name and sorting them in descending order we have:
df_new=df_new.sort_values(['eventid'],ascending=False)
df_four=df_new['eventid']
df_four.head()
```

#### As It can be seen from above DataFrame The first four rows (Top four Countries ) that rank in Global Terrorism.

### So the Countries that were most affected in descending order are:

    1. Iraq

    4. India

### Now let's see the frequency of terror attacks according to the months:

```python
# plotting the months vs number of attacks
df.groupby(['imonth']).size().plot(kind='bar')
plt.rcParams["figure.figsize"] = (20,10)
plt.xlabel('Month number----->')
plt.ylabel('number of attacks------>')
```

### The Above Graph tells us that there is no Correlation between The number of Attacks and The month of the year. Terror attacks happen regardless os the month.

#### And also we can notice a small value for zero (zeroth month) This is because the data contains some error where the column of the df contain zero as a value.

#### So to remove all the rows containing zeros(0) we write this following code✈


```python
df_new_filtered=df[df['imonth']>=1]
```

```python
df_new_filtered.head()
```

```python
# plotting the new filtered data with months on X-axis and number of attacks on Y-axis
df_new_filtered.groupby(['imonth']).size().plot(kind='bar',color='g')
plt.rcParams["figure.figsize"] = (20,10)
plt.xlabel('Month number----->')
plt.ylabel('number of attacks------>')
```

#### And there goes the zeroth month bar. Now our Visualization looks more appealing and cleaner.

### Let's see what is the most used method in a terror attack


For this visualization we will be using the groupby method from the pandas 

```python
# plotting type of attack vs the number of attacks
df.groupby(['attacktype1_txt']).size().plot(kind='bar')
plt.rcParams["figure.figsize"] = (20,10)
plt.xlabel('type of attack----->')
plt.ylabel('number of attacks------>')
```

#### SO Bombing/ Explosions are the first choice of Terrorists and the second is Armed Assualt.

### lets see the main targets of these attacks

```python
# plotting possible targets for attacks
df.groupby(['targtype1_txt']).size().plot(kind='bar',color='m')
plt.rcParams["figure.figsize"] = (20,10)
plt.xlabel('type of attack----->')
plt.ylabel('number of attacks------>')
plt.xticks(rotation=45)
```

#### Private Citizens and property are the most unprotected and susseptible to attacks and hence are attacked more than anybody, and next on the list are Military and Buisnesses.
